```python
import easyocr
import cv2 
import matplotlib.pyplot as plt
```

```python
reader = easyocr.Reader(['en'], gpu=False)
```

```python
image = cv2.imread('street.jpg') 
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()
```

```python
image_data = reader.readtext(image)
```

```python
print(image_data[0])
```

```python
for datum in image_data:
  coordinates, text, confidence_level = datum 
  top_left = (int(coordinates[0][0]), int(coordinates[0][1]))
  bottom_right = (int(coordinates[2][0]), int(coordinates[2][1]))
  cv2.rectangle(image, top_left, bottom_right, (0, 255, 40), 10)
```

```python
plt.imshow(image)
plt.show()
```

```python
cv2.imwrite('output.jpg', image)
```
